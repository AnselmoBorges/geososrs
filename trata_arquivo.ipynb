# Tratando o arquivo Abrigos RS
Esse notebook visa tratar o arquivo para posteriormente ser usado na plotagem do mapa

```python
import pandas as pd
```

```python

df = pd.read_csv('planilha-rs.csv')


import requests
import json

def get_lat_long(endereco):
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address={}&key=<COLOQUE_SUA_API_AQUI>'.format(endereco)
    response = requests.get(url)
    data = json.loads(response.text)
    if data['status'] == 'OK':
        return data['results'][0]['geometry']['location']
    return None

for index, row in df.iterrows():
    endereco = row['endereco']
    lat_long = get_lat_long(endereco)
    if lat_long:
        df.at[index, 'latitude'] = lat_long['lat']
        df.at[index, 'longitude'] = lat_long['lng']
    else:
        print('Erro ao buscar lat long para o endereco:', endereco)

df.to_csv('../planilha-rs.csv', index=False)

print('Planilha atualizada com sucesso!')
```

# Pega cidade com longitude e latitude
Com base na longitude e latitude usa a API do Google pra pegar a cidade de cada linha

```python
df = pd.read_csv('planilha-rs.csv')

import googlemaps
gmaps = googlemaps.Client(key='<COLOQUE_SUA_API_AQUI>')

def load_data():
    data = pd.read_csv('planilha-rs.csv')
    return data

def get_city(latitude, longitude):
    reverse_geocode_result = gmaps.reverse_geocode((latitude, longitude))
    if reverse_geocode_result:
        address_components = reverse_geocode_result[0]['address_components']
        for component in address_components:
            if 'locality' in component['types']:
                return component['long_name']
            elif 'administrative_area_level_2' in component['types']:
                return component['long_name']
    return "Localidade Desconhecida"

# Carregar dados
data = load_data()

# Aplicar a função get_city para cada linha do DataFrame (isso pode demorar dependendo do número de linhas)
data['Cidade'] = data.apply(lambda x: get_city(x['LATITUDE'], x['LONGITUDE']), axis=1)

# Gravar a nova versão da planilha com a coluna cidade
data.to_csv('planilha-rs.csv', index=False)
```
